```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Display max columns in dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.shape
```

### Exploratory analytics

1. Missing values
2. 

```python
# Features with NA values
features_with_na = [feature for feature in data.columns if data[feature].isnull().sum() > 1]
features_with_na   

# these are the features having null values
```

```python
# Lets print the name and percentage of features with null values
for feature in features_with_na:
    print('{0} - Null value count: {1} - Percentage: {2}'.format(feature, data[feature].isnull().sum(), 
                                                                 np.round((data[feature].isnull().sum()) / (data[feature].isnull().count()) * 100, 3)))
```

```python
# Analyze missing values and dependent values - Relationship between missing values and Sales price

########   OBS    ######### 
# Here we've converted the Null values to 1 and not-null values to 0. 
# And later we've observed that, median sales price is high, where more number of Null values are present. 
# Due to nan values significance role on Sales price, so we need to replace it with some meaningful data. 

for feature in features_with_na[:]:
    temp_data = data.copy()
#     print(temp_data.columns)
    # if the observation is missing - replace with 1
    temp_data[feature] = np.where(temp_data[feature].isnull(), 1, 0)
    
    # Calculate mean salesprice where information is missing or present
    temp_data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### List of numerical Variables

```python
# here are the list of numerical features
numerical_features = [feature for feature in data.columns if  data[feature].dtype != 'object']

print('Number of numerical features: {}'.format(len(numerical_features)))
```

```python
# Print column names and their data types

# for x in data.columns:
#     print('Name: {}, type {}'.format(x, data[x].dtype))
```

```python
# Visualize numerical features
data[numerical_features].head()
```

```python
# List of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
# Explore the content of these year variables
for feature in year_feature:
    print('Name', feature, '- Number of unique features: ', len(set(data[feature])), '- Unique features', 
          data[feature].unique())
```

```python
# Check relation between house sold year and sale price

plt.figure(figsize=(8, 6))
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median sale price')
plt.show()
```

```python
# Compare different year features with sale price
for feature in year_feature:
    if feature != 'YrSold':
        temp_data = data.copy()
        
        # difference between year feature and year sold
        temp_data[feature] = temp_data['YrSold'] - temp_data[feature]
        
        plt.scatter(temp_data[feature], temp_data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

    We can clearly see from above scatter plots that, when year of built and garage year built is increasing, 
    then the price is decreasing

    Within numerical variables - It's Discrete and Continous variables

```python
discrete_features = [feature for feature in numerical_features if len(data[feature].unique()) < 25 and 
                    feature not in year_feature+['Id']]
print('Length of discrete_features: ', len(discrete_features))
```

```python
discrete_features
```

```python
# Find relationship between discrete features and sale price

for feature in discrete_features:
    temp_data = data.copy()
    temp_data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title('Feature and Sale price')
    plt.show()
```

### Continuous feature

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+
                       year_feature+['Id']]
print('Number of Continuos features: ', len(continuous_features))
```

```python
continuous_features
```

```python
# Since features are continuous, so plot histogram
# We can observe that some features are not in normal distribution

for feature in continuous_features:
    temp_data = data.copy()
    temp_data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title('Histogram for '+ feature)
    plt.show()
```
